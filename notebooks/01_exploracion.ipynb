## 2. Carga del conjunto de datos

En esta sección se carga el archivo CSV original ubicado en la carpeta `data/raw`. Además, se verificará la codificación del archivo para asegurar que caracteres especiales como tildes y la letra "ñ" se lean correctamente.

```python
from pathlib import Path

ruta = Path("../data/raw/3._Accidentes_de_Transito_ocurridos_en_el_Municipio_de_Bucaramanga_20260704.csv")

print(ruta.exists())
```

```python
encodings = [
    "utf-8",
    "latin-1",
    "cp1252"
]

for encoding in encodings:
    print(f"\nProbando encoding: {encoding}")

    try:
        df = pd.read_csv(ruta, encoding=encoding)

        print("Lectura exitosa")
        print(df.columns.tolist())

    except Exception as e:
        print(f"Error: {e}")
```

### Conclusión

Se probaron tres codificaciones comunes para archivos CSV (`utf-8`, `latin-1` y `cp1252`).

La codificación **UTF-8** fue la única que permitió leer correctamente los nombres de las columnas y los caracteres especiales (tildes y la letra `ñ`).

Por esta razón, el resto del proyecto utilizará `encoding="utf-8"` para cargar el conjunto de datos.

## 3. Carga del conjunto de datos
